```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.inflationtool.com/rates/indonesia/historical"
data = requests.get(url).text
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
print('Classes of each table:')
for table in soup.find_all('tr'):
    print(table.get('class'))
```

```python
tables = soup.find_all('table')

table = soup.find('table', class_='text-sm border dataTable-table')
```

```python
df = pd.DataFrame(columns=['Year', 'Jan', 'Feb',	'Mar',	'Apr',	'May',	'Jun',	'Jul', 'Aug', 'Sep',	'Oct',	'Nov',	'Dec',	'Avg.'])

for row in tables.tbody.find_all('tr'):
    columns = row.find_all('td')

    if(columns != []):
        Year = columns[0].text.strip()
        Jan = columns[1].text.strip()
        Feb = columns[2].text.strip()
        Mar = columns[3].text.strip()
        Apr = columns[4].text.strip()
        May = columns[5].text.strip()
        Jun = columns[6].text.strip()
        Jul = columns[7].text.strip()
        Aug = columns[8].text.strip()
        Sep = columns[9].text.strip()
        Oct = columns[10].text.strip()
        Nov = columns[11].text.strip()
        Dec = columns[12].text.strip()
        Avg = columns[13].text.strip()

        df = df.append({'Year': Year, 'Jan':Jan, 'Feb':Feb,	'Mar':Mar,	'Apr':Apr,	'May':May,	'Jun':Jun,	'Jul':Jul, 'Aug':Aug, 'Sep':Sep,	'Oct':Oct,	'Nov':Nov,	'Dec':Dec,	'Avg.':Avg})
```
